**Task 1: Basic Text processing Pipeline**

```python
import nltk
```

```python
text1= '''Natural language processing (NLP) is a subfield of linguistics, 
computer science, information engineering, and artificial intelligence 
concerned with the interactions between computers and human (natural) languages, 
in particular how to program computers to process and analyze large amounts 
of natural language data. '''

for txt in text1:
  sen=nltk.sent_tokenize(text1)
  print(sen)
  for sentence in sen:
    word = nltk.word_tokenize(sentence)
    tag=nltk.pos_tag(word)
    print(tag)
```
